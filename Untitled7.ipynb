##ECN_2024

```python
import pandas as pd

df_cod = pd.read_excel('Encuesta Clima de negocios 2024 final anonimizada.xlsx', sheet_name='Cod')
display(df_cod.head())
```

```python
selected_columns = ['P37_8', 'P56_1', 'P58', 'P62', 'P63', 'F3', 'FACTOR_REGION', 'FACTOR_TOTAL']

for col in df_cod.columns:
    if col.startswith('P60') or col.startswith('P61A') or col.startswith('P61B'):
        selected_columns.append(col)

df_selected = df_cod[selected_columns]
df_selected.drop(columns = ['P61A_OTRO', 'P61A_OTRO_COD','P61B_OTRO', 'P61B_OTRO_COD', 'P61B_9'], inplace = True)

df_selected = df_selected[df_selected['F3'] == 1].drop(columns=['F3'])

display(df_selected.head())
```

'P37_8' es binaria \\
'P56_1' seria la variable a predecir


```python

df_selected['P37_8'] = df_selected['P37_8'].fillna(0)
df_selected['P37_8'] = df_selected['P37_8'].replace(8, 1)
df_selected['P37_8'] = df_selected['P37_8'].astype(int)

df_selected['P60'] = df_selected['P60'].fillna(0)
df_selected['P60'] = df_selected['P60'].replace(2, 0)
df_selected['P60'] = df_selected['P60'].astype(int)

df_selected['P58'] = df_selected['P58'].replace(2,0)


display(df_selected.head())
```

```python
for col in df_selected.columns:
    if col.startswith('P61A'):
        df_selected[col] = df_selected[col].fillna(0)
        df_selected[col] = df_selected[col].replace({1:1, 2:1, 3:1, 4:1, 5:1, 6:1, 7:1, 8:1, 90:1})
        df_selected[col] = df_selected[col].astype(int)

    if col.startswith('P61B'):
        df_selected[col] = df_selected[col].fillna(0)
        df_selected[col] = df_selected[col].replace({1:1, 2:1, 3:1, 4:1, 5:1, 6:1, 7:1, 8:1, 90:1})
        df_selected[col] = df_selected[col].astype(int)

display(df_selected.head())

```

```python
df_selected.to_csv('ecn2024.csv', index=False)
```

* P37 mejor seguridad: 1 o 0
* P56_1 escala de seguridad: 1,2,3,4,5
* P58 sufrio delito: 1 o 0
* P63: edad
* Factores de expansión
* Denuncio el delito: 1 o 0
* P61A medidas de prevencion del delito
* P61B costos de la inseguridad
